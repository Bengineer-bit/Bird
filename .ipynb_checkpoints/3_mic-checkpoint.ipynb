```python
import os
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
from scipy.signal import spectrogram
from IPython.display import Audio
from scipy import signal

# Microphone positions
mic1_initial = np.array([-0.06, 0])  # Mic 1 at [-0.3, 0]
mic2_initial = np.array([0.06, 0])   # Mic 2 at [0.3, 0]
mic3_initial = np.array([0.03, 0])     # Mic 3 at [0, 0]
mic4_initial = np.array([-0.03, 0]) 
mic_distance = np.linalg.norm(mic2_initial - mic1_initial)  # Distance between mic 1 and mic 2
midpoint = (mic1_initial + mic2_initial + mic3_initial + mic4_initial) / 4  # Midpoint between mic 1 and mic 2

# Sound source position
sound_source = np.array([2, 1])  # Initial sound source position for alignment

# Angle from midpoint to sound source
angle_diff = -np.arctan2(sound_source[0] - midpoint[0], sound_source[1] - midpoint[1])

# Function to rotate microphones and calculate alignment loss
def mic_rotation(angle_to_rotate, sound_source, midpoint):
    mic1_rotated = midpoint + (mic_distance / 2) * np.array([np.cos(angle_to_rotate), np.sin(angle_to_rotate)])
    mic2_rotated = midpoint + (-mic_distance / 2) * np.array([np.cos(angle_to_rotate), np.sin(angle_to_rotate)])
    mic3_rotated = np.array([0, 0])  # Mic 3 remains stationary at origin
    return mic1_rotated, mic2_rotated, mic3_rotated

mic1_rotated, mic2_rotated, mic3_rotated = mic_rotation(angle_diff, sound_source, midpoint)

# Plotting the initial and aligned microphone positions
plt.figure(figsize=(8, 8))
plt.plot([mic1_initial[0], mic2_initial[0]], [mic1_initial[1], mic2_initial[1]], 'bo-', label="Initial Mic Positions")
plt.plot(mic3_initial[0], mic3_initial[1], 'bo', label="Initial Mic 3", markersize=8)
plt.plot([mic1_rotated[0], mic2_rotated[0]], [mic1_rotated[1], mic2_rotated[1]], 'ro-', label="Aligned Mic Positions")
plt.plot(mic3_rotated[0], mic3_rotated[1], 'ro', label="Aligned Mic 3", markersize=8)
plt.plot(sound_source[0], sound_source[1], 'gx', label="Sound Source", markersize=10)
plt.plot(midpoint[0], midpoint[1], 'k+', label="Midpoint")
plt.xlabel("X Position (m)")
plt.ylabel("Y Position (m)")
plt.title("2D Sound Localization with Perpendicular Microphones")
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()

# Constants for loss calculation
SOUND_SPEED = 343.2  # Speed of sound (m/s)

# Display the final rotated microphone positions
print("Rotated Microphone Positions (Perpendicular):")
print("Mic 1:", mic1_rotated)
print("Mic 2:", mic2_rotated)
print("Mic 3:", mic3_rotated)
```

```python
# Loss calculation function for a range of angles
def loss_angle(angle_to_rotate):
    mic1_rotated, mic2_rotated, mic3_rotated = mic_rotation(angle_to_rotate, sound_source, midpoint)

    # Distances to microphones
    dist_to_mic1 = np.linalg.norm(sound_source - mic1_rotated)
    dist_to_mic2 = np.linalg.norm(sound_source - mic2_rotated)
    dist_to_mic3 = np.linalg.norm(sound_source - mic3_rotated)

    t_to_mic1 = dist_to_mic1 / SOUND_SPEED
    t_to_mic2 = dist_to_mic2 / SOUND_SPEED
    t_to_mic3 = dist_to_mic3 / SOUND_SPEED

    samples_at_mic1 = int(t_to_mic1 * SR)
    samples_at_mic2 = int(t_to_mic2 * SR)
    samples_at_mic3 = int(t_to_mic3 * SR)

    y1 = y[samples_at_mic1:]
    y2 = y[samples_at_mic2:]
    y3 = y[samples_at_mic3:]

    min_length = min(len(y1), len(y2), len(y3))
    y1 = y1[:min_length]
    y2 = y2[:min_length]
    y3 = y3[:min_length]

    sos = signal.butter(1, 500, 'lp', fs=SR, output='sos')
    y1 = signal.sosfilt(sos, y1)
    y2 = signal.sosfilt(sos, y2)
    y3 = signal.sosfilt(sos, y3)

    loss = np.linalg.norm(y1 - y2) + np.linalg.norm(y1 - y3) + np.linalg.norm(y2 - y3)
    return loss

angles = np.linspace(-np.pi / 2, np.pi / 2, 101)
losses = np.array([loss_angle(a) for a in angles])

sos = signal.butter(1, 4, 'lp', fs=1 / (np.pi / 101), output='sos')
smoothed_losses = signal.sosfilt(sos, losses)

plt.figure()
plt.plot(np.degrees(angles), losses, label="Loss")
plt.plot(np.degrees(angles), smoothed_losses, label="Smoothed Loss")
plt.xlabel("Angle (degrees)")
plt.ylabel("Loss")
plt.title("Loss vs. Rotation Angle with 3 Microphones")
plt.legend()
plt.show()


```
